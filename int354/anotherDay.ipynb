```python
from sklearn.datasets import load_iris
dataset = load_iris()
x = dataset.data
y = dataset.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
x_val,x_test,y_val,y_test=train_test_split(x_test,y_test,test_size=0.5)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.pipeline import Pipeline
```

```python
p = Pipeline([('sc',StandardScaler()),('pca1',PCA(n_components=2)),('p',Perceptron(penalty='l2'))])
```

```python
p.fit(x_train,y_train)
```

```python
train_pred1 = p.predict(x_train)

val_pred1 = p.predict(x_val)

from sklearn.metrics import accuracy_score
print("Training Accuracy: ",accuracy_score(train_pred1,y_train))
print("Validation Accuracy: ",accuracy_score(val_pred1,y_val))
```

```python
p1 = Pipeline([('sc',StandardScaler()),('pca1',PCA(n_components=3)),('p',Perceptron(penalty='l2'))])
```

```python
p1.fit(x_train,y_train)
```

```python
train_pred = p1.predict(x_train)

val_pred = p1.predict(x_val)
print("Training Accuracy: ",accuracy_score(train_pred,y_train))
print("Validation Accuracy: ", accuracy_score(val_pred,y_val))  
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA 
p3 = Pipeline([('sc',StandardScaler()),('lda',LDA(n_components=2)),('p',Perceptron(penalty='l1'))])
```

```python
p3.fit(x_train,y_train)
```

```python
train_pred4 = p3.predict(x_train)
val_pred4 = p3.predict(x_val)
print("Training Accuracy: ",accuracy_score(train_pred4,y_train))
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5,shuffle=True)
```

```python
acc_score = []
```

```python
for train_index, val_index in kf.split(x_train):
    x_train1,x_val1 = x_train[train_index],x_train[val_index]
    y_train1,y_val1 = y_train[train_index],y_train[val_index]
    p.fit(x_train1,y_train1)
    val_pred1 = p.predict(x_val1)
    acc = accuracy_score(val_pred1,y_val1)
    acc_score.append(acc)
```

```python
print("Accuracy Score: ",acc_score)
```

```python
#average accuracy
import numpy as np
print("Average Accuracy: ",np.mean(acc_score))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
kf1 = KFold(n_splits=5,shuffle=True)
results = cross_val_score(p,x_train,y_train,cv=kf1)
print("Results: ",results)
```
