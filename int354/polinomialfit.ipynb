```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X = np.array([258.0,270.0,294.0,320.0,342.0,368.0,396.0,446.0,480.0,586.0])[:, np.newaxis]
Y = np.array([236.0,234.0,252.0,298.0,314.0,342.0,360.0,391.2,390.8,312.0,])[:, np.newaxis]
```

```python
lr = LinearRegression() #for simple linear regression
pr = LinearRegression() #for polynomial linear regression

quadratic = PolynomialFeatures(degree=2) #pass the degree of the polynomial you want to fit
```

```python
X_quad = quadratic.fit_transform(X) #transform the input features to higher degree
```

```python
lr.fit(X,Y)
X_fit = np.arange(250,600,10)[:, np.newaxis] #generate data from rnage 350 to 600 stepping by 10, then split each into an array.
y_lin_fit = lr.predict(X_fit)
```

### Fit the multiple linear regression model by considering the polynomial regression

```python
pr.fit(X_quad,Y)
Y_poly_fit = pr.predict(quadratic.fit_transform(X_fit))
```

### plot the results

```python
plt.scatter(X,Y, label='Training Data points')
plt.plot(X_fit,y_lin_fit,label='Linear Fit',linestyle='--')
plt.plot(X_fit,Y_poly_fit,label='Quadratic Fit')
plt.legend(loc='upper left')
plt.show();
```

```python
y_lin_pred = lr.predict(X) # linear regression
y_quad_pred = pr.predict(X_quad) #quadratice reg
print('Training MSE of linear model: %3f, Quadratic : %.3f' %(mean_squared_error(Y,y_lin_pred),mean_squared_error(Y,y_quad_pred)))
```

```python
print('Training R^2 of linear model : %.3f , Quadratic : %.3f' %(r2_score(Y,y_lin_pred),(r2_score(Y,y_quad_pred))))
```
